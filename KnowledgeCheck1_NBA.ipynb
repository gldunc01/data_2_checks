```python
from urllib import response
import requests
import pandas as pd
import json
import numpy as np
import config
```

## 1) PULL IN DATA FROM AN API.

```python
#Importing Data from API
url='https://api-nba-v1.p.rapidapi.com/players'

querystring = {"team":"2","season":"2021"}

headers = {
	"X-RapidAPI-Key": config.api_key,
	"X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

response.status_code
```

```python
json = response.json()
json
```

```python
json.keys()
```

```python
#Pulling in the data we want from the json
data = json['response']
data
```

```python
#Converting to dataframe
df = pd.DataFrame(data)
df
```

## DATA CLEANING

```python
#Normalizing fields that have dictionaries nested
births = pd.json_normalize(df.birth)
births

nba = pd.json_normalize(df.nba)
nba

height = pd.json_normalize(df.height)
height

weight = pd.json_normalize(df.weight)
weight

leagues = pd.json_normalize(df.leagues)
leagues
```

```python
#Combining the newly normalized fields into one df
dfs= pd.concat([births, nba, height, weight, leagues], axis=1)
dfs
```

```python
#Merging original dataframe with normalized dataframe
df_normalized = df.merge(dfs, how='inner', right_index=True, left_index=True)
df_normalized
```

```python
#Removing unnecessary columns
df_normalized = df_normalized.drop([
    'nba','height', 'weight', 'leagues', 'birth',
    'vegas.jersey','vegas.active', 'vegas.pos', 
    'utah.jersey', 'utah.active', 'utah.pos', 
    'africa.jersey', 'africa.active', 'africa.pos',
    'standard.jersey','standard.active', 'standard.pos', 
    'sacramento.jersey', 'sacramento.active', 'sacramento.pos'
    ], axis=1)
df_normalized
```

```python
#Column Renaming
df_normalized = df_normalized.rename(columns=
    {
    'date':'birthdate', 
    'country': 'birthcountry',
    'start':'nba_startdate',
    'pro': 'years_pro',
    'feets': 'height_ft',
    'inches':'height_inch',
    'meters': 'height_meters',
    'pounds':'weight_lbs',
    'kilograms': 'weight_kg'
})
df_normalized
```

```python
#Changing data types
df_normalized['birthdate'] = pd.to_datetime(df_normalized['birthdate'])

df_normalized = df_normalized.fillna(value=np.nan)

df_normalized.replace(to_replace=['None'], value=np.nan, inplace=True)
df_normalized
```

```python
df_normalized['height_ft'] = df_normalized['height_ft'].astype(float).astype("Int32")
df_normalized['height_inch'] = df_normalized['height_inch'].astype(float).astype("Int32")
df_normalized['height_meters'] = df_normalized['height_meters'].astype(float)
df_normalized['weight_lbs'] = df_normalized['weight_lbs'].astype(float).astype("Int32")
df_normalized['weight_kg'] = df_normalized['weight_kg'].astype(float)

df_normalized.dtypes
```

```python
#Converting the height to be in one field instead of 2
df_normalized['converted_height'] = ((df_normalized['height_ft'])*12) + ((df_normalized['height_inch'])/12)
df_normalized
```
